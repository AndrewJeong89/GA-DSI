```python
import spotipy
sp = spotipy.Spotify()

def extract_id(artist, title):
    results = sp.search(q=artist+" "+title) 
    for i, t in enumerate(results['tracks']['items']):
        return t['id']
```

```python
ids = []
for x in tks.index:
    ids.append(extract_id(tks["Artist"][x], tks["Title"][x]))
ids=np.array(ids)
```

```python
tks['id'] = ids
```
