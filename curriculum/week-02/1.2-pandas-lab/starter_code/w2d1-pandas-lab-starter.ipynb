```python
# Create a histogram by creating a list comprehension + a function to turn the data into inches
```

```python

def height(x):
    if x == 0:
        return 0
    
    elif 'Under 5 ft.'in x:
        return 58
    
    elif '6\'6" and above' in x:
        return 78
    
    else:
        ft_in = int(x.split("'")[0]*12)
        inc = int(x.split("'")[1].replace('"',''))
        inches = ft_in + inc
        return inches

char['tall'] = [height(x) for x in char['tall']]
```

```python
char['tall'].hist(figsize=(10,8),color='k',bins= 20)
```
