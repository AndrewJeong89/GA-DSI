```python
import pandas as pd
import numpy as np
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.cross_validation import train_test_split
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

data_2015 = pd.read_csv('/Users/andrewjeong/Downloads/airbnb_data.csv',index_col=0)
data_2015.rename(columns = {'neighbourhood':'neighborhood'},inplace = True)
data_2015.head()
```

```python
cols = ['month','bedrooms','number_of_reviews','review_scores_rating','neighborhood','num_comp','occ%','avg_price']
data = data_2015.loc[:,cols]
data['neighborhood'] = le.fit_transform(data['neighborhood'])

X1 = data.iloc[:,:5]
y1 = data['avg_price']
X1_train,X1_test,y1_train,y1_test = train_test_split(X1,y1, random_state = 25,test_size=0.33)
```

```python
gb1 = gb1.fit(X1_train,y1_train)
month = 1
bedrooms = 2
number_of_reviews = 80
review_scores_rating = 80
neighborhood = 10
item = np.array([month, bedrooms, number_of_reviews, review_scores_rating, neighborhood])
gb1_pred = gb1.predict(item)
```

```python
price = gb1_pred[0]
price
```
