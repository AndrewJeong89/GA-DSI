```python
df['Users'] = df['Users'].str.strip()
df['Transactions'] = df['Transactions'].str.strip()
df['Dollars'] = df['Dollars'].str.strip()
```

```python
df['Users'] = df['Users'].str.replace(',', '')
df['Transactions'] = df['Transactions'].str.replace(',', '')
df['Dollars'] = df['Dollars'].str.replace(',', '')
```

```python
df[pd.isnull(df).any(axis = 1)]
```

```python
df = df.dropna()
```

```python
df['Users'] = df['Users'].astype(int)
df['Transactions'] = df['Transactions'].astype(int)
df['Dollars'] = df['Dollars'].astype(float)
df['Week Starting'] = pd.to_datetime(df['Week Starting'])
```

## Trends & Conclusions

As a Cab driver, I would most likely want to work for Yellow Cab in New York City, NY, as there are more users and better ASP and revenue and in New York City, NY, the average ASP and total revenue is highest in this particular
location (higher than overall average US ASP)

Dollars, Transactions, and Users are highly correlated


1. Yellow Cab makes more than 3x more money than Pink Cab in 2014
2. Yellow Cab has more than double the amount of transactions as Pink Cab
3. Yellow Cab has 1.3 million more users than Pink Cab
4. While Pink Cab doesn't have as many users, Pink Cab brings in more revenue per user


1. California and New York brought in the most revenue for both companies because of its city hub areas
2. Average ASP per week were highest in New York, Florida, and Texas compared to the average for the US


1. New York City, San Francisco, Chicago, Los Angeles, and Washington DC brought in the highest revenues
2. New York City, Silicon Valley, Orange County, Miami, Austin, Dallas have the highest mean ASP and higher than than of the average for US


1. The highest revenues in 2014 were from the end of the year as opposed to the beginning of the year
        December had the highest revenue amounts per week

### Data Bias

1. Data bias comes from a disporportionate amount of Users, Transactions and Dollar sales come from company "Yellow Cab" compared to "Pink Cab"
2. Not surprisingly, cities and states with higher populations tend to have higher revenue amounts
        Also, many outside of city hubs tend not to use cabs/ taxis
        Boundaries of Cities/ States
3. There is colinearity between Users, Transactions, and Dollars resulting in overlapping conclusions
        ASP is also a result of Dollars/ Transactions (dervied calculated column)
4. Credit card transactions vs Cash transactions (not included)
5. Smartphone payers
6. Tips


### Correct Bias

1. % of payments in cash vs credit
2. How many people tip and how its included in the data
